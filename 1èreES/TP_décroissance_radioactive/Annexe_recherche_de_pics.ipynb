# La recherche de pics avec seuil sous Python

Il peut être délicat de récupérer des pics à partir d'un graphe

Cette annexe montre comment le faire assez facilement en utilisant la librairie scipy
## 1. Importation des librairies

Bien sûr, avant d'écrire la moindre ligne de code, il faut importer ses librairies :

```python
import numpy as np # ajout de la librairie numpy pour fabriquer des données aléatoires
import matplotlib.pyplot as plt # ajout de la librairie matplotlib.pyplot pour fabriquer les graphes
from scipy.signal import find_peaks # ajout de la librairie scipy.signal pour récupérer la fonction find_peaks()
```

## 2. Initialisation des données

### Initialisation du générateur pseudo-aléatoire de données
Le générateur de nombres np.random est un générateur "pseudo-aléatoire". Concrètement, cela signifie qu'en utilisant les mêmes paramètres de génération, le générateur obtient exactement les mêmes listes de nombres.

La variable "seed" permet d'utiliser une autre liste de nombre. Pour s'en apercevoir, vous pouvez donner la valeur 0, 1, 2, 3 etc. à cette variable et relancer l'éxécution du bloc code suivant :

```python
seed = 0
np.random.seed(seed)
r1 = np.random.randint(0, 200, 20)
r2 = np.random.randint(0, 20, 100)
r = np.concatenate((r1, r2))
print(r)
```

On constate qu'en effet, si on relance le générateur avec une valeur "seed" particulière, on obtient la liste de nombres correspondant à cette valeur.

## 3. Génération des données

```python
np.random.shuffle(r) # utilisation de la fonction shuffle() pour échanger des valeurs dans la liste de nombres r
peaks, _= find_peaks(r, height= 75) # utilisation de la fonction find_peaks() pour rechercher les index de nombres supérieurs à 75
plt.plot(r) # Création du graphe de chaque valeur de r en fonction de sa position dans la liste
plt.plot(peaks, r[peaks], "x") # Ajout des pics dans le graphe
plt.show() # Affichage du graphe
```
